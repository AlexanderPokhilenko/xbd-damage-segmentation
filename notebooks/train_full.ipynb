# Full training run: 4 segmentation models on xBD damage

Trains U-Net, BAFUNet, FusionUNet, ResUNet++ with the same ResNet34 ImageNet-pretrained encoder.
Each run is saved into `checkpoints/<run_name>/` with `last.pt`, `best.pt`, `history.json`, `curves.png`.

Resume is supported: rerunning the cell with the same config will pick up from `last.pt`.

```python
import sys
from pathlib import Path

# When running from notebooks/, add project root to path
ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT))

import os
os.chdir(ROOT)
print(f"Working dir: {Path.cwd()}")
```

```python
import subprocess

CONFIGS = [
    "configs/unet.yaml",
    "configs/bafunet.yaml",
    "configs/fusion_unet.yaml",
    "configs/resunet_pp.yaml",
]

for cfg in CONFIGS:
    print(f"\n{'='*70}\nTraining: {cfg}\n{'='*70}")
    result = subprocess.run(
        ["python", "-m", "src.training.train", "--config", cfg, "--resume", "auto"],
        cwd=str(ROOT),
    )
    if result.returncode != 0:
        print(f"!! {cfg} failed with code {result.returncode}")
        break
```

```python
from pathlib import Path
import pandas as pd

from src.training.plots import build_summary_table, plot_compare_runs

run_names = ["unet_resnet34", "bafunet_resnet34", "fusion_unet_resnet34", "resunet_pp_resnet34"]
run_dirs = [Path("checkpoints") / r for r in run_names]

rows = build_summary_table(run_dirs)
df = pd.DataFrame(rows)
df = df.sort_values("val_dice", ascending=False).reset_index(drop=True)
df
```

```python
from src.training.plots import plot_compare_runs
out_dir = Path("results")
out_dir.mkdir(exist_ok=True)

plot_compare_runs(run_dirs, out_dir / "compare_val_dice.png", metric="dice", split="val",
                  title="Validation Dice across models")
plot_compare_runs(run_dirs, out_dir / "compare_val_iou.png", metric="iou", split="val",
                  title="Validation IoU across models")
plot_compare_runs(run_dirs, out_dir / "compare_val_loss.png", metric="loss", split="val",
                  title="Validation Loss across models")

# Display
from PIL import Image
import matplotlib.pyplot as plt
for p in [out_dir / "compare_val_dice.png", out_dir / "compare_val_iou.png", out_dir / "compare_val_loss.png"]:
    img = Image.open(p)
    plt.figure(figsize=(10, 5))
    plt.imshow(img)
    plt.axis("off")
    plt.show()
```

```python
df.to_csv(out_dir / "main_comparison.csv", index=False)
print(df.to_string(index=False))
```
